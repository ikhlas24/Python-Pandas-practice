2. What is the average score of anime produced by 'Studio Ghibli'?  

```python
studio_ghibli_anime = df[df['Studios'].str.contains('Studio Ghibli', na=False)]

# Calculate the average score, excluding 'UNKNOWN' values
average_score_ghibli = studio_ghibli_anime['Score'][studio_ghibli_anime['Score'] != 'UNKNOWN'].astype(float).mean()

# Print the average score
print(f"The average score of Studio Ghibli anime is: {average_score_ghibli}")
```

3. How many unique anime titles start with the letter 'A'?

4. Which year has the highest number of anime releases?

```python
df['Aired'] = pd.to_datetime(df['Aired'], errors='coerce')

df['Year'] = df['Aired'].dt.year

anime_releases_by_year = df.groupby('Year').size()

anime_releases_by_year.idxmax()
```

5. Which genre has the highest average number of episodes?

```python
# Split genres and expand so each genre is in its own row
anime_genres = df.assign(Genres=df['Genres'].str.split(',')).explode('Genres')

# Convert the 'Episodes' column to numeric, handling errors for non-numeric values (e.g., 'Unknown')
anime_genres['Episodes'] = pd.to_numeric(anime_genres['Episodes'], errors='coerce')

# Group by 'Genres' and calculate the average number of episodes
average_episodes_per_genre = anime_genres.groupby('Genres')['Episodes'].mean()

# Find the genre with the highest average number of episodes
genre_with_highest_avg_episodes = average_episodes_per_genre.idxmax()
highest_avg_episodes = average_episodes_per_genre.max()

# Print the result
genre_with_highest_avg_episodes
```
